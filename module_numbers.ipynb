```python
from numbers import Number
import itertools

MOD = 7


def prime(number):
    """Проверка на простоту"""
    i = 2
    while i * i <= number:
        if number % i == 0:
            return False
        i += 1
    return True


class Modular_numbersDomainError(ValueError):
    pass


class FieldError(ValueError):
    pass


class ZeroDivisionError(ValueError):
    pass


class Modular_numbers:
    """Числа по модулю"""

    def __init__(self, arg=0):
        """x - число"""
        if isinstance(arg, Number):
            self.x = arg % MOD
        else:
            raise Modular_numbersDomainError(
                "You are trying to create modular_numbers from " + repr(arg))

    def __str__(self):
        return (str(self.x))

    def __eq__(self, other):
        if isinstance(other, Number):
            other = Modular_numbers(other)

        if isinstance(other, Modular_numbers):
            return self.x == other.x
        else:
            raise Modular_numbersDomainError(
                "Can't say if modular_numbers is equal to " + str(type(other)))

    def __lshift__(self, deg):
        return Modular_numbers(self.x * 10 ** deg)

    def __add__(self, other):
        if isinstance(other, Number):
            other = Modular_numbers(other)

        return Modular_numbers((self.x + other.x) % MOD)

    def __radd__(self, other):
        return self.__add__(other)

    def __neg__(self):
        return Modular_numbers(MOD - self.x)

    def __sub__(self, other):
        if isinstance(other, Number):
            other = Modular_numbers(other)

        return self.__add__(other.__neg__())

    def __rsub__(self, other):
        return self.__neg__().__add__(other)

    def __mul__(self, other):
        if isinstance(other, Number):
            other = Modular_numbers(other)

        return Modular_numbers(self.x * other.x)

    def __rmul__(self, other):
        return self.__mul__(other)

    def __divmod__(self, other):
        if isinstance(other, Number):
            other = Modular_numbers(other)

        div = MOD
        if prime(MOD) & (other.x != 0):
            for i in range(MOD):
                if i * self.x == other.x:
                    div = i
                    break
            return Modular_numbers(div), Modular_numbers(self.x % other.x)
        elif other.x == 0:
            raise ZeroDivisionError("You are trying to divide by zero")
        else:
            raise FieldError("You are not in Field!")

    def __floordiv__(self, other):
        return divmod(self, other)[0]

    def __mod__(self, other):
        return divmod(self, other)[1]

    def __rfloordiv__(self, other):
        return divmod(other, self)[0]

    def __rmod__(self, other):
        return divmod(other, self)[1]

    def __rdivmod__(self, other):
        return Modular_numbers(other).__divmod__(self)
```

```python
p1 = Modular_numbers()
p2 = Modular_numbers(1234)
p3 = 3

print(p1, p2)
print(p1 == p2)
print(p1 == 0, p2 == 1)
print(p2)

print(p3 + p2)
print((p3 * (p3-p2) * p3) << 2)

m1 = Modular_numbers(123)
m2 = Modular_numbers(321)
m3 = Modular_numbers(45)

m12 = m1 * m2
print(m12)
m123 = m12 + m3
print(m123)
d, r = divmod(m12, m1)
print(d, r)
d = m123 // m1
r = m123 % m1
print(d, r)

print(3 // Modular_numbers(4))
print(3 % Modular_numbers(44))
```
